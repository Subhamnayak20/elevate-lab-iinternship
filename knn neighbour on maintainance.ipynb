### Encoding

```python
#Encoding
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df.team=le.fit_transform(df.team)
df['team'].unique()
```

```python
le.inverse_transform([0,2,1])
```

```python
df.provider=le.fit_transform(df.provider)
df['provider'].unique()
```

```python
le.inverse_transform([3,0,1,2])
```

```python
ip=df.drop('broken',axis=1)
op=df.broken
```

```python
from sklearn.model_selection import train_test_split as tts
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
```

```python
X_train,X_test,Y_train,Y_test=tts(ip,op,test_size=0.2,random_state=43)
X_train.shape,X_test.shape,Y_train.shape,Y_test.shape
```

```python
#standard scaler
X_train=sc.fit_transform(X_train)
X_test=sc.fit_transform(X_test)
```

```python
#KNN neighbour
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn= KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train,Y_train)
```

```python
pred2=knn.predict(X_test)
pred2
```

```python
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score, confusion_matrix, ConfusionMatrixDisplay
```

```python
ac=accuracy_score(pred2,Y_test)
re=recall_score(pred2,Y_test)
pr=precision_score(pred2,Y_test)
f1=f1_score(pred2,Y_test)

print('accuracy ', ac)
print('recall ', re)
print('precision ', pr)
print('f1_score ', f1)
```

```python
cm1=confusion_matrix(pred2,Y_test)
cm1
```

```python
cmd1=ConfusionMatrixDisplay(cm1)
cmd1.plot()
```
